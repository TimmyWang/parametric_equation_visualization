```python
import vis
import numpy as np
```

## 2D - x controlled by parameter p1 and y controlled by parameter p1

```python
def f1(p1,p2):
    import math
    return math.cos(p1)

def f2(p1,p2):
    import math
    return math.sin(p1)

v = vis.Visual(f1,f2)

range_p1 = np.random.normal(0,100,100)
range_p2 = [1] # can be whatever value since neither f1 nor f2 takes value in p2 into account 
v.plot(range_p1, range_p2, figsize=(7,7))
```

## 2D - x controlled by parameter p1 and y controlled by parameter p2

```python
def f1(p1,p2):
    import math
    return math.cos(p1)

def f2(p1,p2):
    import math
    return math.sin(p2)*p2

v = vis.Visual(f1,f2)

range_p1 = np.random.uniform(0,10,200)
range_p2 = np.random.uniform(0,100,50)
v.plot(range_p1, range_p2, figsize=(7,7))
```

## 2D - x controlled by parameter p1, p2 and y controlled by parameter p1

```python
def f1(p1,p2):
    import math
    return math.cos(p1)*p2

def f2(p1,p2):
    import math
    return math.sin(p1)

v = vis.Visual(f1,f2)

range_p1 = np.random.normal(0,100,50)
range_p2 = np.random.normal(0,100,50)
v.plot(range_p1, range_p2, figsize=(12,7))
```

## 2D - x controlled by parameter p1, p2 and y controlled by parameter p1, p2 - case 1

```python
def f1(p1,p2):
    import math
    return math.cos(p1+p2)

def f2(p1,p2):
    import math
    return math.sin(p1-p2)

v = vis.Visual(f1,f2)

range_p1 = np.random.normal(0,10,50)
range_p2 = np.random.normal(0,10,50)
v.plot(range_p1, range_p2, figsize=(12,7))
```

## 2D - x controlled by parameter p1, p2 and y controlled by parameter p1, p2 - case 2

```python
def f1(p1,p2):
    return p1*p2

def f2(p1,p2):
    import math
    if p1 > p2:
        return p1-p2
    else:
        return p1+p2

v = vis.Visual(f1,f2)

range_p1 = np.random.normal(0,10,100)
range_p2 = np.random.normal(0,10,100)
v.plot(range_p1, range_p2, figsize=(12,7), alpha=0.3)
```

## 3D - x, y, z controlled by parameter p1

```python
def f1(p1,p2,p3):
    import math
    return math.sin(p1)

def f2(p1,p2,p3):
    import math
    return math.cos(p1)

def f3(p1,p2,p3):
    import math
    return p1

v = vis.Visual(f1,f2,f3)

range_p1 = np.random.normal(0,10,1000)
range_p2 = [1]
range_p3 = [1]
v.plot(range_p1, range_p2, range_p3, figsize=(12,7))
```

## 3D - x, y, z controlled by parameter p1, p2

```python
def f1(p1,p2,p3):
    import math
    return math.sin(p1)+math.cos(p2)

def f2(p1,p2,p3):
    import math
    return math.sin(p1)-math.cos(p2)

def f3(p1,p2,p3):
    import math
    return math.sin(p1)*math.cos(p2)

v = vis.Visual(f1,f2,f3)

range_p1 = np.random.normal(0,10,100)
range_p2 = np.random.normal(0,10,100)
range_p3 = [1]
v.plot(range_p1, range_p2, range_p3, figsize=(12,7))
```

## 3D - x, y, z controlled by parameter p1, p2, p3

```python
def f1(p1,p2,p3):
    import math
    return math.sin(p1)+math.cos(p2)+p3

def f2(p1,p2,p3):
    import math
    return math.sin(p1)-math.cos(p2)+p3

def f3(p1,p2,p3):
    import math
    return math.sin(p1)*math.cos(p2)+p3

v = vis.Visual(f1,f2,f3)

range_p1 = np.random.normal(0,10,20)
range_p2 = np.random.normal(0,10,20)
range_p3 = np.random.uniform(0,1,20)
v.plot(range_p1, range_p2, range_p3, figsize=(12,7))
```
